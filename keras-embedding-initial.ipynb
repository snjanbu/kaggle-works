## Importing required libraries

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
sentences = ["I am good", "I can do it", "Nothing is impossible", "Nothing is permanent", "Stay positive"]
```

## Initializing constants

```python
vocabulary_size = 1000
sentence_length = 10
dimension_length = 8
```

```python
encoded_sentences = [one_hot(sentence, vocabulary_size) for sentence in sentences]
padded_sentences = pad_sequences(encoded_sentences, sentence_length)
```

## Constructing Model

```python
model = Sequential()
model.add(Embedding(vocabulary_size, dimension_length, input_length=sentence_length))
model.compile(loss='mse', optimizer='adam', metrics='mse')
```

```python
model.summary()
```

```python
model.predict(padded_sentences)
```
